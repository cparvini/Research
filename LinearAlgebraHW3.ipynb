```python
import numpy
import math
from matplotlib import pyplot
```

```python
# initialize variables
x = numpy.linspace(-math.pi, math.pi, 1000)
n = [2, 5, 10, 20, 40]
n_count = 0
fn_store = numpy.empty((len(x),len(n)))

for N in n:
    A = numpy.empty((N+1))
    B = numpy.empty((N))
    Ni = numpy.arange(1,N)
    fn = numpy.ones((len(x)))
    
    A[0] = (numpy.exp(math.pi) - numpy.exp(-math.pi))/(2*math.pi)
    
    for i in range(1,N):
        A[i] = (2/math.pi) * ((math.sinh(math.pi)*math.cos(math.pi*i) + i*math.cosh(math.pi)*math.sin(math.pi*i))/(i**2 + 1))
        B[i-1] = (2/math.pi) * ((math.cosh(math.pi)*math.sin(math.pi*i) + i*math.sinh(math.pi)*math.cos(math.pi*i))/(i**2 + 1))
    
    fn = fn * A[0]
    
    for j in range(0,N-1):
        fn += A[j+1]*numpy.cos(Ni[j]*x)
        fn += B[j]*numpy.sin(Ni[j]*x)
    
    fn_store[:,n_count] = fn[:]
    n_count += 1
```

```python
fig = pyplot.figure(figsize=(16,8))
fig.suptitle('Trigonometric Fourier Series Values with Varying Number of Terms (n)', fontsize=20)
ax1 = pyplot.plot(x, fn_store[:,0], label='n = 2')
ax2 = pyplot.plot(x, fn_store[:,1], label='n = 5')
ax3 = pyplot.plot(x, fn_store[:,2], label='n = 10')
ax4 = pyplot.plot(x, fn_store[:,3], label='n = 20')
ax5 = pyplot.plot(x, fn_store[:,4], label='n = 40')
pyplot.xlabel('$x$', fontsize=20)
pyplot.ylabel('$f_{n}$', fontsize=20)
pyplot.legend()
pyplot.savefig('FourierN.jpg')
```
