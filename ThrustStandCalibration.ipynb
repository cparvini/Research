```python
import numpy
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# Initialize variables
a = numpy.linspace(1E-3,5E-3,10)
d = 1.5 * a
V = numpy.linspace(1,1000,10)
N = 12
k = 9E9 # Vm/C
q = numpy.zeros((N,1))
z = numpy.zeros((N,1))
zp = numpy.zeros_like(z)
zeta = numpy.zeros_like(z)
zetap = numpy.zeros_like(z)
F = numpy.empty((len(a), len(d), len(V)))
```

```python
# Calculate

print('  a        d      V            F')
print('====================================')

aic = 0
dic = 0
Vic = 0

for ai in a: 
    for di in d:
        for Vi in V:
            q[0] = ai * Vi / k
            z[0] = di + ai
            zp[0] = 1
            zeta[0] = 1
            zetap[0] = 0
            
            for i in range(1,N):
                z[i] = z[0] - ai**2 / (z[0] + z[i-1])
                q[i] = q[i-1] * ai / (z[0] + z[i-1])
                zeta[i] = zeta[i-1] * ai / (z[0] + z[i-1])
                
                zp[i] = 1 + ai**2 * zp[i-1] / (z[0] + z[i-1])**2
                zetap[i] = (ai / (z[0] + z[i-1])) * (-zeta[i-1] * (1 + zp[i-1])/(z[0] + z[i-1]) + zetap[i-1])
                
            F[aic, dic, Vic] = (-ai * Vi**2 / (2 * k)) * numpy.sum(zetap)
            print('%.2g    %.2g    %.2g        %.2g' % (ai, di, Vi, F[aic, dic, Vic]))
            Vic += 1
            
        Vic = 0
        dic += 1
    dic = 0
    aic += 1
```

```python
ref = 9

fig = pyplot.figure(figsize=(20, 8))
fig.suptitle('Contours for d = %.2f and %.2f mm' %(d[0]*1000, d[ref]*1000), fontsize=20)
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

ax1.contourf(a*1000, V/1000, F[:,0,:]*1E6)
ax1.set_xlabel('a (mm)', fontsize=20)
ax1.set_ylabel('V (kV)', fontsize=20)
ax1.set_zlabel('F ( $\mu$ N )', fontsize=20)

ax2.contourf(a*1000, V/1000, F[:,ref,:]*1E6)
ax2.set_xlabel('a (mm)', fontsize=20)
ax2.set_ylabel('V (kV)', fontsize=20)
ax2.set_zlabel('F ( $\mu$ N )', fontsize=20)
```
