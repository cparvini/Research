```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
# Initialize variables
a = numpy.linspace(1E-3,5E-3,5)
d = 4 * a
V = numpy.linspace(1,30,30)
N = 12
k = 9E9 # Vm/C
q = numpy.zeros((N,1))
z = numpy.zeros((N,1))
zp = numpy.zeros_like(z)
zeta = numpy.zeros_like(z)
zetap = numpy.zeros_like(z)
```

```python
# Calculate

print('  a        d      V            F')
print('====================================')

for ai in a:
    for di in d:
        for Vi in V:
            q[0] = ai * Vi / k
            z[0] = di + ai
            zp[0] = 1
            zeta[0] = 1
            zetap[0] = 0
            
            for i in range(1,N):
                z[i] = z[0] - ai**2 / (z[0] + z[i-1])
                q[i] = q[i-1] * ai / (z[0] + z[i-1])
                zeta[i] = zeta[i-1] * ai / (z[0] + z[i-1])
                
                zp[i] = 1 + ai**2 * zp[i-1] / (z[0] + z[i-1])**2
                zetap[i] = (ai / (z[0] + z[i-1])) * (-zeta[i-1] * (1 + zp[i-1])/(z[0] + z[i-1]) + zetap[i-1])
                
            F = (-ai * Vi**2 / (2 * k)) * numpy.sum(zetap)
            print('%.2g    %.2g    %.2g        %.2g' % (ai, di, Vi, F))
```
